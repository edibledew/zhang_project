```python
import numpy as np
import pandas as pd
from IPython.core.display import display
import matplotlib.pyplot as plt
import matplotlib.cm as cm

data = pd.read_csv('Dataset_Clean.csv')
display(data.info())

import seaborn as sns
sns.set(style='white')

corr = data.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

f, ax = plt.subplots(figsize=(11, 10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap="coolwarm", vmax=1, vmin=-1, center=0,
            square=True, linewidths=.5, cbar_kws={'label': 'Pair-wise correlation value'})

plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("heatmap.png", dpi=300)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Dataset_Clean.csv')


x_var = 'Cocat_1' 
y_var = 'Cocat_2'  
z_var = 'Tot_Gas_Yield_umol_gcat_h'      

# 创建数据透视表，x_var 与 y_var 定义网格，z_var 的平均值填充颜色
pivot_table = data.pivot_table(index=y_var, columns=x_var, values=z_var, aggfunc='mean')

# 绘制热图
plt.figure(figsize=(15, 8))
sns.heatmap(pivot_table, cmap='Blues', annot=True, linewidths=1, cbar_kws={'label': 'Average Yield'})
plt.xlabel(x_var)
plt.ylabel(y_var)
plt.title('Heatmap of Yield by {} and {}'.format(x_var, y_var))
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("yield_heatmap.png", dpi=300)

plt.show()
```

```python
y_datas = list(set(data['Cocat_2'].values.tolist()))
x_datas = list(set(data['Cocat_1'].values.tolist()))

box_array = np.array([0 for a in range(len(y_datas)*len(x_datas))]).reshape(len(y_datas),len(x_datas)).astype("object")

for raw in range(len(y_datas)):
    for column in range(len(x_datas)):
        df = data[(data['Cocat_1'] == x_datas[column]) & (data['Cocat_2'] == y_datas[raw])]
        if len(df.index) == 0:
            #print(x_datas[column], y_datas[raw])
            data_for_raw_column = np.nan
        else:
            data_for_raw_column = df['Tot_Gas_Yield_umol_gcat_h'].mean(axis = 0)
            
            
        box_array[raw, column] = data_for_raw_column
        
data_heatmap = box_array.astype(float)
# 绘制热图
plt.figure(figsize=(15, 8))
sns.heatmap(data_heatmap, cmap='Blues', annot=True, linewidths=1, cbar_kws={'label': 'Average Yield'})
plt.xlabel(x_var)
plt.ylabel(y_var)
plt.title('Heatmap of Yield by {} and {}'.format(x_var, y_var))
plt.xticks(ticks=np.arange(len(x_datas)), labels=x_datas, rotation=45)
plt.yticks(ticks=np.arange(len(y_datas)), labels=y_datas,rotation=0)
plt.tight_layout()
plt.savefig("yield_heatmap.png", dpi=300)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('Dataset_Clean.csv')

ref = pd.read_csv('periodictable.csv', index_col = 0)
ref_dict = ref["period"].to_dict()

data["Co-catalyst 1_2"] = data["Cocat_1"].replace(to_replace = ["NiO", "CPDs", "MgO", "AgBr", "ln2O3", "Ru (II) dinuclear complex ", "Bi2S3", "CuO", "No"], 
                              value =["Ni", "None", "Mg", "Ag", "In", "Ru", "Bi", "Cu", "None"])

data["Co-catalyst 1_2"] = ["Ru" if x is "Ru (II) dinuclear complex " else x for x in data["Co-catalyst 1_2"]]


period_list = [ref_dict.get(x) for x in data["Co-catalyst 1_2"]]

period_list = [0 if x is None else x for x in period_list]

period_list

#classification_dict = {"NiO", "Pd", "Ag", "Cu", "Au", "Pt", "Co", "V", "Cr", "Ru", "", "", ""}
```
